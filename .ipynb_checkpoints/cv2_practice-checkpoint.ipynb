counters ---- counting the objects

```python
img1= cv2.imread(os.path.join('.','birds_1.jfif'))
```

```python
plt.imshow(img1)
```

```python
gray=cv2.cvtColor(img1,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
ret,thresh=cv2.threshold(gray,125,255,cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh,cmap='gray')
```

```python
counters,hierarchy=cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
for cnt in counters:
    print(cv2.contourArea(cnt))
```

```python
for cnt in counters:
    if cv2.contourArea(cnt) >150:
        cv2.drawContours(img1,cnt,-1,(0,255,0),1)
        x1,y1,x2,y2=cv2.boundingRect(cnt)
        cv2.rectangle(img1,)
    
```

```python
plt.imshow(img1)
```

```python
for cnt in counters:
    if cv2.contourArea(cnt):
        x1,y1,w,h=cv2.boundingRect(cnt)
        cv2.rectangle(img1,(x1,y1),(x1+w,y1+w),(0,255,0),1)
```

```python
plt.imshow(img1)
```

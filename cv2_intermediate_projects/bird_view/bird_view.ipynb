```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('cards_img.jfif')
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
pt1= np.float32([[275,146],[421,59],[445,343],[598,242]])
```

```python
pt1
```

```python
point_imag = cv2.circle(img_rgb, (int(pt1[0][0]), int(pt1[0][1])), 7, (0, 255, 0), cv2.FILLED)
point_imag = cv2.circle(img_rgb, (int(pt1[1][0]), int(pt1[1][1])), 7, (0, 255, 0), cv2.FILLED)
point_imag = cv2.circle(img_rgb, (int(pt1[2][0]), int(pt1[2][1])), 7, (0, 255, 0), cv2.FILLED)
point_imag = cv2.circle(img_rgb, (int(pt1[3][0]), int(pt1[3][1])), 7, (0, 255, 0), cv2.FILLED)
```

```python
plt.imshow(point_imag)
```

```python
w,h=250,350
```

```python
pt2= np.float32([[0,0],[w,0],[0,h],[w,h]])
```

```python
metrics=cv2.getPerspectiveTransform(pt1,pt2)
```

```python
print(metrics)
```

```python
output= cv2.warpPerspective(img_rgb,metrics,(w,h))
```

```python
plt.imshow(output)
```
